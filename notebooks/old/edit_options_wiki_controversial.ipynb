```python
import copy
import matplotlib.pyplot as plt
import numpy as np
import operator
import os
import pandas as pd
import pprint
import scipy.stats
import scipy.spatial.distance
import scipy.special

from valuebank import *
os.chdir('..')
```

```python
df = pd.read_json('data/wiki_controversial_small.jsonl', lines=True)
```

```python
no_rephrase = df[~df['rephrase']]
question_topics = list(zip(no_rephrase['question'], no_rephrase['topic']))
```

```python
options = extract_answers(question_topics, 'gpt-4', count_tokens=False)
```

```python
import itertools
expanded_opts = list(itertools.chain(*[[opt] * 3 for opt in options]))
```

```python
df['options'] = expanded_opts
```

```python
# The below with the full data frame
```

```python
df = df[df['options'].notnull()]
```

```python
unique_option_stances = df.apply(lambda x: len(x['options'].values()) == len(set(x['options'].values())), axis=1)
df = df[unique_option_stances]
```
